```python
import pandas as pd
import requests

dfs = []

for i in range(1, 533):
    response = requests.get(
        f"https://api.themoviedb.org/3/movie/top_rated"
        f"?api_key=6a0c86df9f64ec487967253670fbadfe"
        f"&language=en-US&page={i}"
    )

    data = response.json()

    if 'results' not in data:
        continue

    temp_df = pd.DataFrame(data['results'])[
        ['id', 'title', 'overview', 'popularity',
         'release_date', 'vote_average', 'vote_count']
    ]

    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```
